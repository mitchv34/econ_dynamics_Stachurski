**Example:**
The following algorithm computes the maximum of a function $f$ over a finite set $S$ using a for loop and prints it to the screen.

<div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">

```

````python
# The same algorithim implemented in Julia

# Function
f(x) = x^3-6*x^2 + 1

# Set of points
S = -3.5:0.24:5


plot(f, -4, 7, label="Function", legend=:bottomright)
scatter!(S, f.(S), markersize=3, label="Points to evaluate")
````

````python
# algorithim
c = -Inf
for x in S
    c = maximum([c, f(x)])
end
print(c)
````

**Exercise 2.1.2**
The statemet a ∉ B  may require coding at a lower level. Rewrite the algorithm with an inner loop that steps through each b in B, testing whether a = b, and setting subset = False if no matches occur.


````python
# answer 2.1.2
A = [1,2,3]
B = [1,2,3,4]

subset = true

for a ∈ A
    for b ∈ B
        subset = a == b
        if subset break end
    end
    if ~subset break end
end

println(subset) 
````

**Exercise 2.1.3** Consider a game that pays $1 if three consecutive heads occur in ten flips and zero otherwise. Modify the previous algorithm to generate a round of the game and print the payoff.

````python
# Answer 2.1.3

p = .5
count = 0

for i ∈ 1:10
    U = rand()
    if U < p
        count += 1
    else
        count = 0
    end
    if count >= 3
        break
    end
end

payoff = ( count >= 3 ) ? 1 : 0
````
